Сколько выбросов найдёт классический метод межквартильного размаха в признаке SkinThickness?

```python
diabetes.SkinThickness.describe()

#Для начала необходимо вычислить первый и третий квартили (Q1 и Q3) и межквартильный размах (IQR):
Q1 = np.percentile(diabetes['SkinThickness'], 25)
Q3 = np.percentile(diabetes['SkinThickness'], 75)
IQR = Q3 - Q1

#Затем можно определить нижнюю и верхнюю границы выбросов:
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

#Наконец, можно пройти по всем значениям признака SkinThickness и проверить, не выходят ли они за пределы определенных границ:
outliers = []
for value in diabetes.SkinThickness:
    if value < lower_bound or value > upper_bound:
        outliers.append(value)
print(len(outliers))
```

```python
# импортируем библиотеку scipy для использования теста Шапиро-Уилка
from scipy.stats import shapiro

# проверяем распределение признака SkinThickness на нормальность
stat, p = shapiro(diabetes['SkinThickness'])
print('Statistics=%.3f, p=%.3f' % (stat, p))

# если p-value > 0.05, то распределение признака является нормальным

# вычисляем среднее значение и стандартное отклонение
mean = diabetes['SkinThickness'].mean()
std = diabetes['SkinThickness'].std()

# вычисляем границы выбросов
upper_bound = mean + 3 * std
lower_bound = mean - 3 * std

# определяем выбросы
outliers = diabetes[(diabetes['SkinThickness'] > upper_bound) | (diabetes['SkinThickness'] < lower_bound)]

# выводим выбросы
print(len(outliers))
```

Найдите сначала число выбросов в признаке DiabetesPedigreeFunction с помощью классического метода межквартильного размаха.



```python
# вычисляем межквартильный размах
Q1 = diabetes['DiabetesPedigreeFunction'].quantile(0.25)
Q3 = diabetes['DiabetesPedigreeFunction'].quantile(0.75)
IQR = Q3 - Q1

# вычисляем границы выбросов
upper_bound = Q3 + 1.5 * IQR
lower_bound = Q1 - 1.5 * IQR

# определяем выбросы
outliers = diabetes[(diabetes['DiabetesPedigreeFunction'] > upper_bound) | (diabetes['DiabetesPedigreeFunction'] < lower_bound)]

# выводим выбросы
print(len(outliers))
```

Затем найдите число выбросов в этом же признаке в логарифмическом масштабе (при логарифмировании единицу прибавлять не нужно!). Какова разница между двумя этими числами (вычтите из первого второе)?

```python
# применяем логарифмическое преобразование к признаку DiabetesPedigreeFunction
log_feature = np.log(diabetes.DiabetesPedigreeFunction)

# Находим выбросы в логарифмическом масштабе
Q1 = np.percentile(log_feature, 25)
Q3 = np.percentile(log_feature, 75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = [x for x in log_feature if x < lower_bound or x > upper_bound]
num_outliers = len(outliers)
print(num_outliers)
```
